# 1. Importando bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web

#!pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
# yf.pdr_override()
```

# Obtendo os dados do mercado

```python
#ibov = web.get_data_yahoo('VALE3.SA', start='2010-05-03', end='2013-03-01')
data = yf.download('VALE3.SA', start='2010-05-03', end='2013-03-01')
```

```python
data.head()
```

```python
data.tail()
```

```python
data['MA21'] = data['Open'].rolling(21).mean()
data['MA200'] = data['Open'].rolling(200).mean()
data[['Open','MA21','MA200']].plot(label='VALE',figsize=(22,6))
```

```python
data_fatiado = data[data.index.year == 2011]
data_fatiado[['Open']].plot(label='VALE',figsize=(22,6))
plt.legend()
```

```python
data_fatiado = data[(data.index.year >= 2008) & (data.index.year <= 2015)]
data_fatiado["Close"].plot(figsize=(22,8), label="Tick")
plt.legend()
```

```python
data_fatiado = data[data.index.month == 12]
data_fatiado.tail(50)
```

```python
data_fatiado = data[data.index.day == 14]
data_fatiado.tail(50)
```
