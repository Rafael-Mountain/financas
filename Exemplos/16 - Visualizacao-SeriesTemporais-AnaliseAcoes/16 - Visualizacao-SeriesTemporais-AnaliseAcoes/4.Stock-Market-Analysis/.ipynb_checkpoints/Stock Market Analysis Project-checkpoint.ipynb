## Daily Percentage Change (Variação percentual diária)
Primeiro será calculada a variação percentual diária. A variação percentual diária é definida pela seguinte fórmula:

$ r_t = \frac{p_t}{p_{t-1}} -1$ ou
$r_t = \frac{p_t - p_{t-1}}{p_{t-1}}$ ou com logaritmo
$r_t = \ln \left(\frac{p_t}{p_{t-1}}\right)$



Isso define r_t (retorno no momento t) como igual ao preço no momento t dividido pelo preço no momento t-1 (o dia anterior) menos 1. Basicamente, isso apenas informa sobre seu ganho (ou perda) percentual se você comprou a ação no dia e depois o vendeu no dia seguinte. Embora isso não seja necessariamente útil para tentar prever os valores futuros das ações, é muito útil para analisar a volatilidade das ações. Se os retornos diários tiverem uma distribuição ampla, a ação fica mais volátil de um dia para o outro. Serão calculados os retornos percentuais e, em seguida, será plotado em um histograma para decidir qual ação é a mais estável.

** Crie uma nova coluna para cada dataframe chamada retornos (returns). Esta coluna será calculada a partir da coluna Preço de fechamento (close). Existem duas maneiras de fazer isso, ou um cálculo simples usando o método .shift() que segue a fórmula acima, ou você também pode usar o método pct_change integrado do pandas. **

## Cumulative Daily Returns

Agora pode-se ver qual ação teve a maior variação nos retornos diários (Tesla, o gráfico original do preço das ações também deixou isso óbvio).

Com retornos acumulados diários, a pergunta a responder é a seguinte: se eu investisse $ 1 na empresa no início da série temporal, quanto valeria hoje? Isso é diferente de apenas o preço da ação no dia atual, pois levará em consideração os retornos diários. Tenha em mente que nosso cálculo simples aqui não levará em consideração ações que devolvem dividendos. Vejamos alguns exemplos simples:

Digamos que exista uma ação 'ABC' que está sendo ativamente negociada em bolsa. ABC tem os seguintes preços correspondentes às datas indicadas

    Date                        Price
    01/01/2018                   10
    01/02/2018                   15
    01/03/2018                   20
    01/04/2018                   25

**Daily Return** : O retorno diário é o lucro/perda obtido pela ação em comparação com o dia anterior. (Isso é o que acabamos de calcular acima). Um valor acima de um indica lucro, da mesma forma, um valor abaixo de um indica perda. Também é expresso em porcentagem para melhor transmitir a informação. (Quando expresso em porcentagem, se o valor estiver acima de 0, a ação deu lucro ou perda). Portanto, para o exemplo acima, os retornos diários seriam

    Date                         Daily Return                  %Daily Return
    01/01/2018                 10/10 =  1                          -   
    01/02/2018                 15/10 =  3/2                       50%
    01/03/2018                 20/15 =  4/3                       33%
    01/04/2018                 25/20 =  5/4                       20%

**Cumulative Return**: Embora os retornos diários sejam úteis, não dão ao investidor uma visão imediata dos ganhos que obteve até à data, especialmente se a ação for muito volátil. O retorno acumulado é calculado em relação ao dia em que o investimento é feito. Se o retorno cumulativo for superior a um, você está obtendo lucros, caso contrário, terá prejuízo. Portanto, para o exemplo acima, os ganhos cumulativos são os seguintes

    Date                       Cumulative Return         %Cumulative Return
    01/01/2018                  10/10 =  1                         100 %   
    01/02/2018                  15/10 =  3/2                       150 %
    01/03/2018                  20/10 =  2                         200 %
    01/04/2018                  25/10 =  5/2                       250 %

A fórmula para um retorno diário acumulado é:

$ i_i = (1+r_t) * i_{t-1} $

Aqui podemos ver que estamos apenas multiplicando nosso investimento anterior em i em t-1 por 1 + nossos retornos percentuais. O Pandas torna isso muito simples de calcular com seu método cumprod(). Usando algo da seguinte maneira:

    df[daily_cumulative_return] = ( 1 + df[pct_daily_return] ).cumprod()
    

O método cumprod() é usado para calcular o produto cumulativo de uma série. Ele percorre os valores em uma série (ou DataFrame) e, para cada elemento, multiplica-o pelo valor do elemento anterior, resultando em uma série onde o último elemento contém o produto de todos os valores anteriores
<br>
https://www.w3schools.com/python/pandas/ref_df_cumprod.asp
<br>
https://acervolima.com/numpy-cumprod-em-python

** Crie uma coluna de retorno diário acumulado para o dataframe de cada montadora.**

```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
```

```python
tesla.head()
```

```python
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
gm['Cumulative Return'] = (1 + gm['returns']).cumprod()
```

** Agora, plote as colunas de retorno acumulado contra o índice da série temporal. Qual ação apresentou o maior retorno para um investimento de $1? E qual apresentou o menor?**

```python
tesla['Cumulative Return'].plot(label='Tesla',figsize=(16,8),title='Cumulative Return')
ford['Cumulative Return'].plot(label='Ford')
gm['Cumulative Return'].plot(label='GM')
plt.legend()
```
