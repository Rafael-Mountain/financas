# Rolling and Expanding

A very common process with time series is to create data based off of a rolling mean. Let's show you how to do this easily with pandas!

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Best way to read in data with time series index!
# parse_dates:converter automaticamente colunas de datas em objetos do tipo DateTime
df = pd.read_csv('time_data/walmart_stock.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

Now let's add in a rolling mean! This rolling method provides row entries, where every entry is then representative of the window. 

```python
# 7 day rolling mean
df.rolling(7).mean().head(20)
```

```python
df['Open'].plot()
# fazendo a mdia mvel de 30
df.rolling(window=30).mean()['Close'].plot()
```

Easiest way to add a legend is to make this rolling value a new column, then pandas does it automatically!

```python
# acrescentando a coluna do dataframe
df['Close: 30 Day Mean'] = df['Close'].rolling(window=30).mean()
# plotando os preos close e a mdia mvel de 30
df[['Close','Close: 30 Day Mean']].plot(figsize=(16,6))
```

## expanding

Agora, e se voc quiser levar em conta tudo desde o incio da srie temporal como um valor contnuo? Por exemplo, no considere apenas um perodo de 7 dias, ou mdia mvel mensal, mas em vez disso, leve em considerao tudo desde o incio da srie temporal, continuamente:

```python
# Optional specify a minimum number of periods
# plotando a mdia dos closes em relao a todos os pontos anteriores.
df['Close'].expanding(min_periods=1).mean().plot(figsize=(16,6))
```

## Bollinger Bands

Mais ser falado sobre grficos de anlise financeira e indicadores tcnicos, mas aqui vale a pena mencionar:

Mais informaes: http://www.investopedia.com/terms/b/bollingerbands.asp

*Desenvolvidas por John Bollinger, Bollinger Bands so bandas de volatilidade colocadas acima e abaixo de uma mdia mvel. A volatilidade baseia-se no desvio padro, que muda  medida que a volatilidade aumenta e diminui. As bandas alargam-se automaticamente quando a volatilidade aumenta e estreitam quando a volatilidade diminui. Esta natureza dinmica das Bandas de Bollinger tambm significa que podem ser usadas em diferentes ativos com as configuraes padro. Para sinais, as Bandas de Bollinger podem ser usadas para identificar topos e fundos ou para determinar a fora da tendncia.*

*As Bandas de Bollinger refletem a direo com o SMA de 20 perodos e a volatilidade com as bandas superior/inferior. Como tal, podem ser usados para determinar se os preos so relativamente altos ou baixos. De acordo com Bollinger, as bandas devem conter 88-89% da ao do preo, o que torna significativo um movimento fora das bandas. Tecnicamente, os preos so relativamente elevados quando esto acima da banda superior e relativamente baixos quando esto abaixo da banda inferior. No entanto, relativamente alto no deve ser considerado um sinal de baixa ou de venda. Da mesma forma, relativamente baixo no deve ser considerado uma alta ou um sinal de compra. Os preos so altos ou baixos por um motivo. Tal como acontece com outros indicadores, as Bandas de Bollinger no devem ser usadas como uma ferramenta independente. *

```python
df['Close: 30 Day Mean'] = df['Close'].rolling(window=20).mean()
df['Upper'] = df['Close: 30 Day Mean'] + 2*df['Close'].rolling(window=20).std()
df['Lower'] = df['Close: 30 Day Mean'] - 2*df['Close'].rolling(window=20).std()
df[['Close','Close: 30 Day Mean','Upper','Lower']].plot(figsize=(16,6))
# plot de 4 curvas
```

For expanding operations, it doesn't help very much to visualize this against the daily data, but instead its a good way to get an idea of the "stability" of a stock. This idea of stability and volatility is something we are going to be exploring heavily in the next project.
