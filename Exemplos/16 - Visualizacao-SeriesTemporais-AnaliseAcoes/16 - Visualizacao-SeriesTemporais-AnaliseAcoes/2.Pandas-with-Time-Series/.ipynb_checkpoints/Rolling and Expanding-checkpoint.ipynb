# Rolling and Expanding

A very common process with time series is to create data based off of a rolling mean. Let's show you how to do this easily with pandas!

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Best way to read in data with time series index!
# parse_dates:converter automaticamente colunas de datas em objetos do tipo DateTime
df = pd.read_csv('time_data/walmart_stock.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

Now let's add in a rolling mean! This rolling method provides row entries, where every entry is then representative of the window. 

```python
# 7 day rolling mean
df.rolling(7).mean().head(20)
```

```python
df['Open'].plot()
# fazendo a média móvel de 30
df.rolling(window=30).mean()['Close'].plot()
```

Easiest way to add a legend is to make this rolling value a new column, then pandas does it automatically!

```python
# acrescentando a coluna do dataframe
df['Close: 30 Day Mean'] = df['Close'].rolling(window=30).mean()
# plotando os preços close e a média móvel de 30
df[['Close','Close: 30 Day Mean']].plot(figsize=(16,6))
```

## expanding

Agora, e se você quiser levar em conta tudo desde o início da série temporal como um valor contínuo? Por exemplo, não considere apenas um período de 7 dias, ou média móvel mensal, mas em vez disso, leve em consideração tudo desde o início da série temporal, continuamente:

```python
# Optional specify a minimum number of periods
# plotando a média dos closes em relação a todos os pontos anteriores.
df['Close'].expanding(min_periods=1).mean().plot(figsize=(16,6))
```

## Bollinger Bands

Mais será falado sobre gráficos de análise financeira e indicadores técnicos, mas aqui vale a pena mencionar:

Mais informações: http://www.investopedia.com/terms/b/bollingerbands.asp

*Desenvolvidas por John Bollinger, Bollinger Bands® são bandas de volatilidade colocadas acima e abaixo de uma média móvel. A volatilidade baseia-se no desvio padrão, que muda à medida que a volatilidade aumenta e diminui. As bandas alargam-se automaticamente quando a volatilidade aumenta e estreitam quando a volatilidade diminui. Esta natureza dinâmica das Bandas de Bollinger também significa que podem ser usadas em diferentes ativos com as configurações padrão. Para sinais, as Bandas de Bollinger podem ser usadas para identificar topos e fundos ou para determinar a força da tendência.*

*As Bandas de Bollinger refletem a direção com o SMA de 20 períodos e a volatilidade com as bandas superior/inferior. Como tal, podem ser usados ​​para determinar se os preços são relativamente altos ou baixos. De acordo com Bollinger, as bandas devem conter 88-89% da ação do preço, o que torna significativo um movimento fora das bandas. Tecnicamente, os preços são relativamente elevados quando estão acima da banda superior e relativamente baixos quando estão abaixo da banda inferior. No entanto, relativamente alto não deve ser considerado um sinal de baixa ou de venda. Da mesma forma, relativamente baixo não deve ser considerado uma alta ou um sinal de compra. Os preços são altos ou baixos por um motivo. Tal como acontece com outros indicadores, as Bandas de Bollinger não devem ser usadas como uma ferramenta independente. *

```python
df['Close: 30 Day Mean'] = df['Close'].rolling(window=20).mean()
df['Upper'] = df['Close: 30 Day Mean'] + 2*df['Close'].rolling(window=20).std()
df['Lower'] = df['Close: 30 Day Mean'] - 2*df['Close'].rolling(window=20).std()
df[['Close','Close: 30 Day Mean','Upper','Lower']].plot(figsize=(16,6))
# plot de 4 curvas
```

For expanding operations, it doesn't help very much to visualize this against the daily data, but instead its a good way to get an idea of the "stability" of a stock. This idea of stability and volatility is something we are going to be exploring heavily in the next project.
