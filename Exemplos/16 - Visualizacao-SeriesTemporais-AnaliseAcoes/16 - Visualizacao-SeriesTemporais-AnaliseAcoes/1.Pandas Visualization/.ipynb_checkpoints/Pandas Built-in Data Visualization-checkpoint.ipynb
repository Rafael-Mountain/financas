# Pandas Built-in Data Visualization

Neste notebook serão apresentados os recursos integrados do pandas para visualização de dados. Ele é integrado ao matplotlib, mas foi incorporado ao pandas para facilitar o seu uso.


## Imports

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

## The Data

There are some fake data csv files you can read in as dataframes:

```python
df1 = pd.read_csv('df1',index_col=0)
df2 = pd.read_csv('df2')
```

## Style Sheets

Matplotlib tem [folhas de estilo](https://matplotlib.org/2.0.2/gallery.html#style_sheets) que você pode usar para deixar seus gráficos um pouco mais bonitos. Essas folhas de estilo incluem plot_bmh,plot_fivethirtyeight,plot_ggplot e muito mais. Eles basicamente criam um conjunto de regras de estilo que seus gráficos seguem. É recomendado usá-los, eles fazem com que todos os seus gráficos tenham a mesma aparência e sejam mais profissionais. Você pode até criar o seu próprio se quiser que todos os gráficos de seu trabalho tenham a mesma aparência (embora seja um pouco tedioso criar).

Veja como usá-los.

**Antes de plt.style.use() seus gráficos ficam assim:**

```python
df1['A'].hist()
```

Call the style:

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Now your plots look like this:

```python
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

Let's stick with the ggplot style and actually show you how to utilize pandas built-in plotting capabilities!

# Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)



https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html

Let's start going through them!

## Area

```python
df2.plot.area(alpha=0.4)
```

## Barplots

```python
df2.head()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## Histograms

```python
df1['A'].plot.hist(bins=50)
```

## Line Plots

```python
df1['A'].plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

## Scatter Plots

https://statorials.org/pt/grafico-de-dispersao/

```python
df1.plot.scatter(x='A',y='B')
```

You can use c to color based off another column value
Use cmap to indicate colormap to use. 
For all the colormaps, check out: http://matplotlib.org/users/colormaps.html

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

Or use s to indicate size based off another column. s parameter needs to be an array, not just the name of a column:

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

## BoxPlots

https://frons.com.br/blog/ferramentas/como-fazer-box-plot/

```python
df2.plot.box() # Can also pass a by= argument for groupby
```

## Hexagonal Bin Plot

Useful for Bivariate Data, alternative to scatterplot:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

____

## Kernel Density Estimation plot (KDE)

* https://pt.wikipedia.org/wiki/Estimativa_de_densidade_kernel

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
